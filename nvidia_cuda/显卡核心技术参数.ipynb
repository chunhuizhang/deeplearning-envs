## memory bus width

- 4090: 384-bit, A100: 5120-bit
    - 内存总线宽度是指GPU（cores）与其显存（GPU内存）之间的数据传输通道的宽度
    - A100 5120 bit： 采用的是HBM2（高带宽内存）技术
    - 内存总线宽度越宽，理论上GPU的内存带宽越高
- 4090 计算得快，A100 内存读取得快；

- 计算GPU的理论带宽（不适用于 hbm 的计算）：
    - A100 (HBM2)
        - deviceQuery：memory clock rate，1215mhz
    - 4090 (GDDR6X)
        - deviceQuery：memory clock rate，10501mhz
$$
\text{理论带宽 (GB/s)}= \frac{内存总线宽度 (bits)×内存速度 (MHz)}{8×1024}
$$


```python
5120*10501/(8*1024)
```
