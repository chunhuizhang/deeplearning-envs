- 家用220v，最大瓦数：
    - 比如微波炉是1000瓦，那么电流就是1000W/220V=4.5A。
    - 基本上家庭中最大的功率并不会超过1万3千瓦，所以家用电流一般都是在63A以下。
- 因为家庭一般的用插座大致有两种，一种是10A的插座，另一种是16A的插座。
    - 16A的插座：热水器，空调；
    - 如果没有的话，则需要家用转接头
